# Training Model

```
# Model Evaluation Function

def evaluate(test,pred):
  TP=0
  TN=0
  FP=0
  FN=0

  index=0
  for i in test.index:
      if pred[index]>=0.5:
          pred[index]=1
      else:
          pred[index]=0

      if(pred[index]==1 and test.get(i)==1):
          TP=TP+1
      elif(pred[index]==1 and test.get(i)==0):
          FP=FP+1
      elif(pred[index]==0 and test.get(i)==1):
          FN=FN+1
      else:
          TN=TN+1
      index=index+1



  precision = TP/(TP+FP)
  recall = TP/(TP+FN)
  accuracy = (TP+TN)/(TP+TN+FP+FN)
  F = 2*((precision*recall)/(precision+recall))
  
  
  print('True Positives = {}\nFalse Positives = {}\nTrue Negatives = {}\nFalse Negatives = {}'.format(TP,FP,TN,FN))
  print('----------------------------------------')
  print('Precision = {}\nRecall = {}\nAccuracy = {}\nF-score = {}'.format(precision,recall,accuracy,F))
  
```

```
#Linear Regression Classifier

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,Y_train)
#print(lm.intercept_)
#lm.coef_

pred_lm = lm.predict(X_test)

print('Results of Linear Regression Classifier:')
print('----------------------------------------')
evaluate(Y_test,pred_lm)
```

```
#Logistic Regression Classifier

from sklearn.linear_model import LogisticRegression
logm = LogisticRegression()
logm.fit(X_train,Y_train)
#print(logm.intercept_)
#logm.coef_

pred_log = logm.predict(X_test)


print('Results of Logistic Regression Classifier:')
print('------------------------------------------')
evaluate(Y_test,pred_log)
```

```
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,Y_train)

pred_dt = dtree.predict(X_test)


print('Results of Decision Tree Classifier:')
print('----------------------------------------')
evaluate(Y_test,pred_dt)
```

```
#KNN Regression Classifier

error = []
from sklearn.neighbors import KNeighborsClassifier

for i in range(11,21):
    knn_i = KNeighborsClassifier(n_neighbors=i)
    knn_i.fit(X_train,Y_train)
    pred_i = knn_i.predict(X_test)
    error.append(np.mean(pred_i!=Y_test))
    
plt.plot(range(11,21),error, color='blue', linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.xlabel("Number of nearest neighbors",fontsize=15)
plt.ylabel("Error",fontsize=15)

plt.show()


knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train,Y_train)
pred_knn = knn.predict(X_test)

print('Results of K-nearest Neighbours Classifier:')
print('----------------------------------------')
evaluate(Y_test,pred_knn)
```
